```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# importing dataset
df = pd.read_csv(r'C:\Users\user\Desktop\python projects\movies.csv')
df.head()
```

```python
# checking for missing data
for col in df.columns:
    missing_values = np.mean(df[col].isnull()) 
    print('{} - {}%'.format(col,missing_values))
```

```python
# from the above result, i is evident that there are some missing values. hence we remove them.
df1 = df.dropna()
df1
```

```python
#checking for null values
for col in df1.columns:
    missing_values = np.mean(df1[col].isnull()) 
    print('{} - {}%'.format(col,missing_values))
```

```python
#datatypes of columns
df.dtypes
```

```python
# dropping name,rating,& genre  column for correlations only as well as duplicates
df2 = df1.drop(columns = ['name','rating','genre',])
df2.drop_duplicates().head()
```

```python
# changing datatypes of budget and gross from float to int
df2['budget'] = df2['budget'].astype('int64')
df2['gross'] = df2['gross'].astype('int64')
df2.dtypes
```

```python
# sorting the gross column
df2 = df2.sort_values(by= ['gross'], ascending = False)
df2
```

```python
#splitting the released column
df3 = df2.copy() #making a copy of df2
df3['released'].apply(str) # converting column to a string
df3[['month','day','released_year','county','state','none']] = df3['released'].str.split(' ', expand = True)# splitting each string to different columns and naming them
df3['state'].fillna(" ", inplace = True) # filling None values in state column with an empty string
df4 =df3.sort_values(by= ['gross'], ascending = False) #sorting by gross values
df4.head()
```

```python
df4['realeased_Country'] = df4['county'] +" "+ df4['state'] #concatinating county and state column
df4 = df4.drop(columns = ['month','day','county','state','none']) #dropping some column
df4
```

```python
#some of the rows of the released column do not follow the month,day, year, country format hence the years of those rows returned none or country or states 
pd.set_option('display.max_rows',None) # to view all the columns
df4['released_year'].sort_values()
```

```python
conditions = [df4['released_year'].eq('(Australia)'),
             df4['released_year'].eq('States)'),
             df4['released_year'].eq(None),
             df4['released_year'].eq('(United')]

choices = [df4['year'],df4['year'],df4['year'],df4['year']]
df4['released_year'] = np.select(conditions,choices,default=df4['released_year'])
```

```python
#scatter plot of budget vs Gross Earnings
plt. scatter(x=df2['budget'], y = df2['gross'])
plt.title(" Budget vs Gross Earnings")
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#checking correlations
df2.corr(method = 'pearson') # this works on only numerical field and its of 3 types. Pearson by default, kendall and spearman
```

```python
corr_matrix = df2.corr(method = 'pearson')
plt.subplots(figsize=(10,6))
sns.heatmap(corr_matrix, annot = True)# this is a rectangular plot of data as color encoded matrix, annot = true means if true, write the data value in each cell
plt.title('Correlation matrix for Numeric Features')
plt.xlabel('Movie features')
plt.ylabel('Movie feature')
plt.show()
```

```python
df2.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
corr_matrix = df2.apply(lambda x: x.factorize()[0]).corr(method='pearson') #factorize(): encode strings as numeric values. it is useful for obtaining a numeric representation of an array when all that matters is identifying distinct variables
plt.subplots(figsize=(10,6))
sns.heatmap(corr_matrix, annot = True)# this is a rectangular plot of data as color encoded matrix, annot = true means if true, write the data value in each cell
plt.title('Correlation matrix for Numeric Features')
plt.xlabel('Movie features')
plt.ylabel('Movie feature')
plt.show()
```

```python
corr_mat = df2.apply(lambda x: x.factorize()[0]).corr()
corr_pairs = corr_mat.unstack() # unstack table
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(kind='quicksort') #sort kind can be quicksort, heapsort or mergesort
sorted_pairs
```

```python
strong_corr = sorted_pairs[abs(sorted_pairs) > 0.5]
strong_corr
```
